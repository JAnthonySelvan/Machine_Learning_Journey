```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

**Load Dataset**

```python
dataset = pd.read_csv('data.csv')
```

**Summarize the Data**

```python
print(dataset.shape)
dataset.head(5)
```

**Find and Replace NA values**

```python
dataset.columns[dataset.isna().any()]
```

```python
dataset['hours'] = dataset['hours'].fillna(dataset['hours'].mean())
```

```python
dataset.columns[dataset.isna().any()]
```

**Segregate Dataset into X & Y**

```python
X = dataset.iloc[:, :-1].values
print(X.shape)
X
```

```python
Y = dataset.iloc[:, -1].values
Y
```

**Splitting Train and Test for Error evaluation**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.9,random_state=0)
print(X_train.shape)
print(Y_train.shape)
```

**Training**

```python
model =  LinearRegression()
model.fit(X_train, Y_train)
```

**Evaluate the Errors for Test values**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,root_mean_squared_error
Y_pred = model.predict(X_test)
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse = root_mean_squared_error(Y_test, Y_pred)
print('MAE:', mae)
print("MSE:",mse)
print("RMSE:",rmse)
```

**Predict Mark for Custom Input**

```python
a=[[9.2,20,0]]
PredictedmodelResult = model.predict(a)
print(PredictedmodelResult)
```
