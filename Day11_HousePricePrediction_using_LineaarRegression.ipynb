**Load DataSet**

```python
dataset = pd.read_csv('dataset.csv')
```

**Summarize DataSet**

```python
print(dataset.shape)
dataset.head()
```

**Visualize Dataset**

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(dataset.area,dataset.price,color='red',marker='*')
```

**Segregate DataSet into Input X & Output Y**

```python
X = dataset.drop("price",axis=1)
X
```

```python
Y = dataset.price
Y
```

**Training DataSet Using LinearRegression**

```python
model = LinearRegression()
model.fit(X,Y)
```

**Predict a price for a Input**

```python
x=40000
LandAreainSqFt=[[x]]
PredictedmodelResult = model.predict(LandAreainSqFt)
print("The Price of {0} Square feet Land is: {1}".format(x,PredictedmodelResult[0]))
```
