### *Load Dataset*

```python
dataset = pd.read_csv("SOCR-HeightWeight.csv")
```

### *Summarize Dataset*

```python
print(dataset.shape)
print(dataset.head(5))
print(dataset.columns)
```

### *Segregate Dataset into Input X & Output Y*

```python
X = dataset[['Height(Inches)']].values
X
```

```python
Y = dataset['Weight(Pounds)'].values
# Y=np.reshape(Y,(-1,1))
Y
```

### *Splitting Dataset for Testing our Model*

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.20,random_state=0)
print(x_train.shape,x_test.shape)
```

### *Training Dataset using Decision Tree*

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
# model = LinearRegression()
model = DecisionTreeRegressor(max_depth=3)
model.fit(x_train, y_train)
```

### *Visualizing Graph*

```python
X_val = np.arange(min(X), max(X), 0.1).reshape(-1,1)
X_val = X_val.reshape(-1, 1)

plt.figure()
plt.scatter(X, Y, color='green')
plt.plot(X_val, model.predict(X_val), color='red')

plt.title('Weight prediction using Decision Tree')
plt.xlabel('Height')
plt.ylabel('Weight')

plt.show()
```

### *Prediction for all test data for validation*

```python
ypred = model.predict(x_test)

from sklearn.metrics import r2_score,mean_squared_error
mse = mean_squared_error(y_test,ypred)
rmse=np.sqrt(mse)
print("Root Mean Square Error:",rmse)
r2score = r2_score(y_test,ypred)
print("R2Score",r2score)
```
